**Data cleaning:** It is a process of removing unwanted data by extracting only wanted data.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Practice/amazon_fires.csv",encoding="ISO-8859-1")
df.head()
```

```python
df.info()
```

```python
#1) Renaming the columns
#   d = {"old_Col":"new_col"........}
#syntax:  df.rename(columns=d,inplace=True)
d = {"ano":"year","mes":"month",	"estado":"state",
     "numero":"no_of_fires","encontro":"date"}
df.rename(columns=d,inplace=True)
df
```

```python
#2)Dropping duplicate rows
#syntax:  df.drop_duplicates(inplace=True)
df.drop_duplicates(inplace=True)
df.shape
```

```python
#3)changing the data present in a column
#syntax:  d = {"old_data":"new_data".......}
#         df["col"] = df["col"].map(d)
#note: The len of dict should be equal to no.of unique values
d = {'Janeiro':"jan", 'Fevereiro':"feb", 'Março':"mar", 'Abril':"apr", 'Maio':"may", 'Junho':"jun", 'Julho':"jul",
       'Agosto':"aug", 'Setembro':"sep", 'Outubro':"oct", 'Novembro':"nov", 'Dezembro':"dec"}
df["month"] = df["month"].map(d)
df["month"].unique()
```

```python
#4)Applying inbuilt functions on a column
#syntax:  df[col] = df[col].str.fname(args)
df["month"] = df["month"].str.upper()
df["month"] = df["month"].str.lower()
df["month"] = df["month"].str.capitalize()
df
```

```python
#5) Removing unwanted data
# df["col"] = df["col"].str.strip("unwanted data")
df["no_of_fires"] = df["no_of_fires"].str.strip(" Fires")
df
```

```python
#6) Handling missing values:
df.dropna(inplace=True)
df.isnull().sum()
df.info()
```

```python
#7)changing the data type of col
#df[col] = df[col].astype(dest_dtype)
df["no_of_fires"] = df["no_of_fires"].astype(float)
df["no_of_fires"] = df["no_of_fires"].astype(int)
df["no_of_fires"].unique()
```

```python
#8) Dropping unwanted columns
#  df = df.drop([col1,col2,...],axis=1)
df = df.drop("state",axis=1)
df
```

```python
#9)dropping unwanted rows
#  df = df.drop([in1,in2,,.....inn])

df = df.drop([0,1,2])
df
```

```python
#10)resetting the index
#  df.reset_index(inplace=True)
df.reset_index(inplace=True)
df = df.drop("index",axis=1)
df
```

```python
df = pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/laptopData.csv")
df.head()
```

```python
df.info()
```

```python
#handling Nan values
df.dropna(inplace=True)
df
```

```python
#Remove Unnamed: 0 col
df = df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
#Drop duplicate rows
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.head()
```

```python
#Handling Inches column
df["Inches"] = df["Inches"].replace("?","0")
df["Inches"] = df["Inches"].astype(float)
avg_inch = round(df["Inches"].mean())
df["Inches"] = df["Inches"].replace(0.0,avg_inch)
df["Inches"].unique()
```

```python
df
```

```python
#handling memory column
df["Memory"].mode()
df["Memory"] = df["Memory"].replace("?","256GB SSD")
df["Memory"].unique()
```

```python
#Handling weight column
df["Weight"] = df["Weight"].str.strip("kg")
df["Weight"] = df["Weight"].replace("?","0")
df["Weight"] = df["Weight"].astype(float)
avg_weight = round(df["Weight"].mean())
avg_weight
df["Weight"] = df["Weight"].replace(0.0,avg_weight)
df["Weight"].unique()
```

```python
df.reset_index(inplace=True)
df = df.drop("index",axis=1)
df
```
