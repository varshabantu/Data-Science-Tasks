```python
#3. Concept
# Indexing and slicing
#1. Indexing : It is the process of fetching a single element from an array
# Syntax:
# 1D -->  arr[index]
# 2D -->  arr[row_index, col_index]
# 3D -->  arr[row_index, col_index, val_index]

# 1D
arr1 = np.arange(1,11)
arr1[5]
arr1[3]
arr1[-1]
# print(arr1[5])

# 2D
arr2 = np.array(([1,2,3],
                 [4,5,6],
                 [6,7,8]))
arr2[1,1]
arr2[2,1]
arr2[0,2]
arr2[2,0]

#3D

arr3 = np.array([[[1,2,3,4],
                  [6,7,8,9]],

                 [[4,4,6,7],
                  [7,8,9,0]]])
arr3[1,0,2]
arr3[0,1,2]
arr3[1,1,2]
```

```python
# Slicing on arrays: It is the process of fetching multiple values from a given array.
# Syntax:
# 1D ---> arr[si:ei+=1:upd/step]
# 2D ---> arr[row_slice, col_slice]
# 3D ---> arr[row_slice, col_slice, val_slice]

# 1D slicing:

arr = np.arange(1,7)
arr[:3]
arr[1:5]
arr[::-1]      # Which will give reverse order
arr[0::2]      # where even index is there i want that values.
arr[1::2]      # odd index position.

# 2D slicing:

arr = np.array([[1,2,3,4],
                [3,4,5,6],
                [6,7,7,8]])
arr[0:1,:]
arr[0:2,0:2]
arr[0:2,2:]
arr[2:,:]     # "::"--> means it will fetch all the records from the row
arr[::2,:2]
arr[:,1::2]



# 3D Slicing

arr = np.array([[[1,2,3,4],
                 [6,7,8,9]],
                [[3,4,5,6],
                 [6,9,8,0]]])
arr[0:1,1:,:]
arr[0:1,:,0:3]
arr[:,:,0:2]
```

```python
#4. Operations on Arrays
# We have to perform in 3 types

#1. Array with array
#2. Array with constants/scalar
#3. Math operations using inbuilt methods


#1. Array with array: Another name of array with array operation is called "Vectorization".

# Element wise operations are performed.
# arr1 operator arr2 (two array operands)

# Rules:-
# 1. We need minimum of two arrays.
# 2. Both arrays should be of same length

# Operators: +,-,*,/,//,%
# Floor division will give only the integers.

arr1 = np.arange(0,6)
arr2 = np.arange(4,10)
print(arr1)
print(arr2)
arr1+arr2
arr1-arr2
arr1*arr2     # It say element wise operations
arr1/arr2
arr1//arr2
arr1%arr2
arr1 & arr2
arr1 | arr2
arr1 ^ arr2
~arr1

arr1.all() and arr2.all()    # If non-default all the values are non- default value then it will return true.If we have one false the final output will be false.
arr1.any() and arr2.any()    # True
not arr1.all()  #True
arr1.all() #False
```

```python
#2. Array with constant/Scalar

#Syntax: arr operator constant/scalar

arr = np.arange(1,6)
print(arr)
arr + 2
2 + arr # It can be operator or constand / constant over array anything can work.
arr - 3
arr * 2.5
arr / 5
arr // 5
arr % 2
arr ** 2
arr & 2
arr << 1
arr >>1
arr | 2
```

```python
#3. Math operations using inbuilt methods

arr = np.arange(1,11)
print(arr)

np.sum(arr)
np.mean(arr)
arr.size
np.min(arr)
np.max(arr)
np.sqrt(arr)
np.log(arr)
np.exp(arr)
np.cos(arr)
np.sin(arr)
np.add(arr,arr)
np.subtract(arr,arr)
np.median(arr)
np.multiply(arr,arr)
np.divide(arr,arr)
```

```python
#4. Broadcasting:- Eventhough the shapes are different vectorization will happen the process is called as broadcasting.

#Rule: Numpy always checks the shape from Right side --->Left.
#1. Same no.of columns should be there (or)
#2. Columns should be 1.
arr1 = np.array([[1,2,3],
                 [4,5,6]])
arr2 = np.array([[1],
                 [2]])
print("arr1: ",arr1.shape)
print("arr2: ",arr2.shape)

arr1 + arr2
```
